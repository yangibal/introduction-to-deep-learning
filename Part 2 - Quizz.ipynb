```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

import tensorflow as tf
keras = tf.keras
layers = keras.layers
models = keras.models
optimizers = keras.optimizers
from sklearn import datasets


iris_data_dict = datasets.load_iris()
```

```python
iris_data_dict
```

```python
iris_features = pd.DataFrame(iris_data_dict["data"],columns=iris_data_dict['feature_names'])
iris_targets = pd.DataFrame(iris_data_dict["target"],columns=["species"])
iris_targets["species"] = iris_targets["species"].apply(lambda x: "setosa" if x == 0 else ("versicolor" if x == 1 else "virginica"))
```

```python
iris_features.head()
```

```python
iris_targets
```

```python
iris_targets = pd.get_dummies(iris_targets['species'])
iris_targets.head()
```

```python
full_set = pd.concat([iris_features, iris_targets], axis=1)
full_set.head()
```

```python
training_dataset = full_set.sample(frac=0.8)
testing_dataset = full_set[~full_set.index.isin(training_dataset.index)]
training_dataset.columns
```

```python
input_columns = [
    'sepal length (cm)', 'sepal width (cm)', 'petal length (cm)',
       'petal width (cm)'
]
```

```python
training_dataset.head()
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
training_dataset[input_columns] = scaler.fit_transform(training_dataset[input_columns])
testing_dataset[input_columns] = scaler.transform(testing_dataset[input_columns])

training_dataset.head()
```

```python
iris_model = models.Sequential([
    layers.Dense(8, input_dim=len(input_columns), activation='relu'),
    layers.Dropout(0.1),
    layers.Dense(8, activation='relu'),
    layers.Dropout(0.1),
    layers.Dense(3, activation='softmax')])

iris_model.compile(loss='categorical_crossentropy', optimizer=optimizers.Adam(), metrics=['accuracy'])

iris_model.summary()
```

```python
targets = ["setosa", "versicolor", "virginica"]
history_iris_model = iris_model.fit(training_dataset[input_columns],training_dataset[targets],epochs=850, validation_split=0.2)
```

```python
test_loss, test_acc = iris_model.evaluate(
    testing_dataset[input_columns], 
    testing_dataset[targets]
)

print(f"Evaluation result on Test Data : Loss = {test_loss}, accuracy = {test_acc}")
```
